```python
#  author：IF  
import pandas as pd
import heapq
from datetime import timedelta####引用时间序列模块  
from datetime import time
import threading
from datetime import datetime

####加锁####加锁####加锁####加锁####加锁####加锁####加锁
c = threading.RLock() #由于Python的datetime和time中的_strptime方法不支持多线程，运行时会报错,故在使用的时候加锁
def f(x):
    with c:
        return datetime.strptime(x,"%Y/%m/%d %H:%M:%S")
    
from math import radians, cos, sin, asin, sqrt
def geodistance(lng1,lat1,lng2,lat2):
    lng1, lat1, lng2, lat2 = map(radians, [float(lng1), float(lat1), float(lng2), float(lat2)]) # 经纬度转换成弧度
    dlon=lng2-lng1
    dlat=lat2-lat1
    a=sin(dlat/2)**2 + cos(lat1) * cos(lat2) * sin(dlon/2)**2 
    distance=2*asin(sqrt(a))*6371*1000 # 地球平均半径，6371km
    distance=round(distance/1000,3)
    return distance
```

```python
Result_1204SC = IC_1204SC_copy.copy()
Result_1204SC.insert(loc=10,column='updown',value=0,allow_duplicates=False)
Result_1204SC.insert(loc=11,column='Station',value='待定',allow_duplicates=False)
Result_1204SC.insert(loc=12,column='id',value=0,allow_duplicates=False)
for x in range(len(Result_1204SC)):
    UP = SXZD[SXZD['线路'].isin([Result_1204SC.所属线路[x]])]
    upzipped = zip(UP.Longitude,UP.Latitude,UP.name,UP.id)
    up = list(upzipped)

    DW = XXZD[XXZD['线路'].isin([Result_1204SC.所属线路[x]])]
    dwzipped = zip(DW.Longitude,DW.Latitude,DW.name,DW.id)
    dw = list(dwzipped)##分别获取上行线路

    upresult = []
    for i in range(len(up)):
        lng1,lat1,lng2,lat2 = (Result_1204SC.Longitude[x],Result_1204SC.Latitude[x],up[i][0],up[i][1])
        upresult.append(geodistance(lng1,lat1,lng2,lat2))
    upid = upresult.index(min(upresult))#最小值所在位置
    # print(min(upresult))

    dwresult = []
    for i in range(len(dw)):
        lng1,lat1,lng2,lat2 = (Result_1204SC.Longitude[x],Result_1204SC.Latitude[x],dw[i][0],dw[i][1])
        dwresult.append(geodistance(lng1,lat1,lng2,lat2))
    dwid = dwresult.index(min(dwresult))#最小值所在位置
    # print(min(dwresult))  

    ###将GPS坐标分别遍历上下行站点，计算各自的GPS距离

    if min(upresult) > min(dwresult):
        GPSupdown = XXZD.updown[dwresult.index(min(dwresult))]#如果上行线路最近站点GPS距离 大于  下行线路最近站点GPS距离 则输出上下行方向
        station = dw[dwid][2]#输出站名
        stationid = dw[dwid][3]
    else:
        GPSupdown = SXZD.updown[upresult.index(min(upresult))]#如果上行线路最近站点GPS距离 大于  下行线路最近站点GPS距离 则输出上下行方向
        station = up[upid][2]#输出站名
        stationid = up[upid][3]
    Result_1204SC.at[x,'updown']=GPSupdown
    Result_1204SC.at[x,'Station']=station
    Result_1204SC.at[x,'id']=stationid
```
